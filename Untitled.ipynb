```python
import numpy as np
```

```python
# Sine wave generator
def sine(sampling_rate, duration, frequency, amplitude=1.0):
    x = np.linspace(0., duration, duration * sampling_rate)
    print(x)
    return amplitude * np.sin(2 * np.pi * frequency / x)
```

```python
s = sine(44100, 2.0, 440)
```
